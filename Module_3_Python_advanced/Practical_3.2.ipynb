Notice that your function finishes with a `print` statement? With functions, we usually prefer to finish with a `return` statement if the function has produced something.

 This allows for flexibility: rather than always printing the output, it can also be stored or passed easily to another function. When only using a `print` statement, any other functions cannot access the outputs of your function. 
 
 Note that the `return` statement by itself will not print anything, and therefore does not produce any visible output. If you want to print the outputs of your function, follow the logic in the cell below!

It's important to know that this return statement also finishes the function, __even if there is still code coming after__! Any code written within a function after a return statement, will simply not be executed.

````python
# This function builds a nested dictionary by prompting the user for all of the information

def create_dictionary():
    
    outer_dict = {}
    add_entry = True

    while add_entry:
        entry = ...
        outer_dict.update({entry: {
            "...": ...,
            "...": ...,
            }
        })

        continue_yes_no = input("Continue yes or no? ")
        if continue_yes_no == "No" or continue_yes_no == "no":
            break

    return outer_dict # here we don't use a print statement anymore!

print(create_dictionary()) # we now print here!
````

#### Opening and closing files

Before we work on the actual ASCII art for today, let's talk about how to open and close files with Python. 

Python has a native function for opening files, simply called `open()`. Please check the official Python documentation on it [here](https://docs.python.org/3/library/functions.html#open). You will find that there are many ways of opening files, called "modes". Take some time to read up on the 7 different modes in the documentation - You don't need to fully understand them, but make sure you know that they exist, and get a rough idea of what their differences are. 


Alright! We've written some example code below, to demonstrate the usage of `open()`. \
The first time you run this code, it will automatically create a file called 'text.txt' in the folder you're working in.

````python
# Create the file
file = open('test.txt', "w") # Using "w" allows us to open the file in "writing" mode

# Write something into it
file.write("Now the file has more content!")

# Close the file again
file.close()
````

If we just want to read the file, we can do so with the following code.

````python
# Open the file
file = open('test.txt', "r") # Using "r" allows us to open the file in "reading" mode

# Print the contents of the file
print(file.read())

# Always close the file when you're done with it!
file.close()
````

However, as you will see, using the reading mode "r", you won't be able to write into the file you opened. 


````python
# Open the file
file = open("test.txt", "r")

# Write to the file
file.write("Now the file has even more content!")

# Close the file
file.close()
````

While the above demonstrate the basic use of opening, reading, and writing files, there is one more thing we would like to show you before you have a go at it. There is actually a way that makes Python close the file for you automatically, just to save lines of code. We do this with the `with` statement. The code below is completely equivalent to the code in the second cell above:


````python
with open("test.txt", "r") as file:
    print(file.read())
````
