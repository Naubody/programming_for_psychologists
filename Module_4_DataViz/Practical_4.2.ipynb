#### Linear regression

There are many libraries that can help you with linear regression, like [scipy](https://scipy.org/), [scikit-learn](https://scikit-learn.org/stable/), [statsmodels](https://www.statsmodels.org/stable/index.html) or even just the Python-native statistics package called ... you guessed it... [statistics](https://docs.python.org/3/library/statistics.html). We will be doing a [simple linear regression](https://en.wikipedia.org/wiki/Simple_linear_regression), meaning that only one explanatory variable will be taken into account. For this exercise, we recommend using scipy, but any of the mentioned packages will work!
